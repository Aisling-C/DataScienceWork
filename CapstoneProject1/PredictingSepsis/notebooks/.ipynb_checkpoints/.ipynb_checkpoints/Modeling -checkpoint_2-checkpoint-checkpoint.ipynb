# Load Libraries & Data

```python
import glob, os
import pandas as pd
import numpy as np
import json
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import Image
from scipy.stats import uniform
from datetime import datetime
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
os.chdir("../src/data")
from time_elapsed import time_elapsed
```

```python
os.chdir("../../data/interim")

#Load in Data
#First, the notes
with open("Data_Notes.json", "r") as infile:  
    notes=json.load(infile)
    
os.chdir("../processed")
#Next, the actual data
df=pd.read_csv('sepsis_data_preprocessed.csv',index_col=0)

os.chdir("../../docs")
Image(filename='SepsisClassifier.PNG') 
```

```python
df.info()
```

```python
#Sklearn imports
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.dummy import DummyClassifier
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.metrics import plot_confusion_matrix, classification_report
from sklearn.feature_selection import SelectFromModel
```

```python
#Test-Train Split
X=df.drop(columns=['SepsisLabel','SepsisOnset', 'Hourly_Sepsis', 'Pre_Sepsis', 'Pre_Acute_Sepsis','id'])
y=pd.concat([df.loc[:,'Hourly_Sepsis':'Pre_Acute_Sepsis'],df.SepsisLabel],axis=1)
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.25,random_state=42)

#Fit Min-Max Scaler to X_train data, transform X_train and X_test data
MM_scale=MinMaxScaler()
MM_scale.fit(X_train)
X_train=MM_scale.transform(X_train)
X_test=MM_scale.transform(X_test)

#Split y_train into various classification columns 
y_train_hr=y_train['Hourly_Sepsis']
y_train_pre_acute=y_train['Pre_Acute_Sepsis']
y_train_pre=y_train['Pre_Sepsis']
y_train_sepsis=y_train['SepsisLabel']

#Split y_test into various classification columns 
y_test_hr=y_test['Hourly_Sepsis']
y_test_pre_acute=y_test['Pre_Acute_Sepsis']
y_test_pre=y_test['Pre_Sepsis']
y_test_sepsis=y_test['SepsisLabel']
```

```python
plt.subplots(figsize=(12,10))
sns.heatmap(df.corr());
```

## Modeling Sepsis Label

### Dummy Classifier

```python
#Model Sepsis Label - Dummy Classifier, Most Frequent
dummy_clf = DummyClassifier(strategy="most_frequent")
y_pred_dummy=dummy_clf.fit(X_train, y_train_sepsis).predict(X_test)
print(classification_report(y_test_sepsis,y_pred_dummy,zero_division=0))
```

### Random Forest

```python
#Find optimal random forest parameters for Modeling Sepsis Label 
#Parameters adapted from : https://towardsdatascience.com/hyperparameter-tuning-the-random-forest-in-python-using-scikit-learn-28d2aa77dd74
start = datetime.now()
RFC = RandomForestClassifier(random_state=42,max_features='sqrt')
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
distributions = {'n_estimators': n_estimators,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
RFC_RS=RandomizedSearchCV(RFC,distributions,random_state=42,n_iter=30,cv=4,n_jobs=-1)
search_RFC_Sepsis=RFC_RS.fit(X_train, y_train_sepsis)
end = datetime.now()
time_elapsed(start,end)
search_RFC_Sepsis.best_params_
```

```python
best_dict=search_RFC_Sepsis.best_params_
```

```python
search_RFC_Sepsis.cv_results_['params']
```

```python
#Determine most important features (based on purity index) in the model
start = datetime.now()
embeded_rf_selector = SelectFromModel(RandomForestClassifier(n_estimators=best_dict['n_estimators'], min_samples_split=best_dict['min_samples_split'],
min_samples_leaf=best_dict['min_samples_leaf'], max_features='sqrt', max_depth=best_dict['max_depth'],random_state=42,n_jobs=-1))
embeded_rf_selector.fit(X_train, y_train_sepsis)
embeded_rf_support = embeded_rf_selector.get_support()
embeded_rf_feature = X.loc[:,embeded_rf_support].columns.tolist()
end = datetime.now()
time_elapsed(start,end)
print(str(len(embeded_rf_feature)), 'selected features')
embeded_rf_feature
```

```python
#Model prediction results for test set
RFC1=RandomForestClassifier(n_estimators=best_dict['n_estimators'], min_samples_split=best_dict['min_samples_split'],
min_samples_leaf=best_dict['min_samples_leaf'], max_features='sqrt', max_depth=best_dict['max_depth'],random_state=42,n_jobs=-1)
RFC1.fit(X_train,y_train_sepsis)
plot_confusion_matrix(RFC1,X_test,y_test_sepsis,cmap=plt.cm.Blues)
plt.show()
print(classification_report(y_test_sepsis,RFC1.predict(X_test)))
```

```python
#Run on training set - clearly overfitting!
plot_confusion_matrix(RFC1,X_train,y_train_sepsis,cmap=plt.cm.Blues)
plt.show()
```

## Modeling Pre-Sepsis

```python
#Same Parameters as best of Sepsis_Label
RFC2=RandomForestClassifier(n_estimators=best_dict['n_estimators'], min_samples_split=best_dict['min_samples_split'],
min_samples_leaf=best_dict['min_samples_leaf'], max_features='sqrt', max_depth=best_dict['max_depth'],random_state=42,n_jobs=-1)
RFC2.fit(X_train,y_train_pre)
plot_confusion_matrix(RFC2,X_test,y_test_pre,cmap=plt.cm.Blues)
plt.show()
print(classification_report(y_test_pre,RFC2.predict(X_test)))
```

```python
#Run on training set - Overfit again
plot_confusion_matrix(RFC2,X_train,y_train_pre,cmap=plt.cm.Blues)
plt.show()
```
