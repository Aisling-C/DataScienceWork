```python
#Sklearn imports
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.dummy import DummyClassifier
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.metrics import plot_confusion_matrix, classification_report, \
roc_curve, plot_precision_recall_curve, precision_recall_curve, confusion_matrix, auc
from sklearn.feature_selection import SelectFromModel
```

```python
#Test-Train Split
X=df.drop(columns=['SepsisLabel','SepsisOnset', 'Hourly_Sepsis', 'Pre_Sepsis', 'Pre_Acute_Sepsis','id'])
y=pd.concat([df.loc[:,'Hourly_Sepsis':'Pre_Acute_Sepsis'],df.SepsisLabel],axis=1)
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.25,random_state=42)

#Fit Min-Max Scaler to X_train data, transform X_train and X_test data
MM_scale=MinMaxScaler()
MM_scale.fit(X_train)
X_train=MM_scale.transform(X_train)
X_test=MM_scale.transform(X_test)

#Split y_train into various classification columns 
y_train_hr=y_train['Hourly_Sepsis']
y_train_pre_acute=y_train['Pre_Acute_Sepsis']
y_train_pre=y_train['Pre_Sepsis']
y_train_sepsis=y_train['SepsisLabel']

#Split y_test into various classification columns 
y_test_hr=y_test['Hourly_Sepsis']
y_test_pre_acute=y_test['Pre_Acute_Sepsis']
y_test_pre=y_test['Pre_Sepsis']
y_test_sepsis=y_test['SepsisLabel']
```

```python
best_dict={'min_samples_split': 4, 'min_samples_leaf': 7, 'max_depth': 81}
```

```python
#Model prediction results for test set
start = datetime.now()
RFC1=RandomForestClassifier(n_estimators=500, min_samples_split=best_dict['min_samples_split'],
min_samples_leaf=best_dict['min_samples_leaf'], max_features='sqrt', max_depth=best_dict['max_depth'],random_state=42,n_jobs=-1)
RFC1.fit(X_train,y_train_sepsis)
end = datetime.now()
time_elapsed(start,end)
plot_confusion_matrix(RFC1,X_test,y_test_sepsis,cmap=plt.cm.Blues)
plt.show()
print(classification_report(y_test_sepsis,RFC1.predict(X_test)))
```

```python
plot_confusion_matrix(RFC1,X_train,y_train_sepsis,cmap=plt.cm.Blues)
plt.show()
```

```python
y_pred_prob_train=RFC1.predict_proba(X_train)
```

```python
y_pred_test=RFC1.predict(X_test)
```

```python
plot_precision_recall_curve(RFC1,X_test,y_test_sepsis)
plt.show()
```

```python
yhat = RFC1.predict_proba(X_test)
yhat = yhat[:, 1]
precision, recall, thresholds = precision_recall_curve(y_test_sepsis, yhat)
fscore = (2 * precision * recall) / (precision + recall)
ix = np.argmax(fscore)
print('Best Threshold=%f, F-Score=%.3f' % (thresholds[ix], fscore[ix]))
```

```python
no_skill = len(y_test_sepsis[y_test_sepsis==1]) / len(y_test_sepsis)
auc_=round(auc(recall,precision),3)
plt.plot([0,1], [no_skill,no_skill], linestyle='--', label='No Skill',c='orange')
plt.scatter(recall, precision, marker='.', label='Random Forest',s=10)
plt.scatter(recall[ix], precision[ix], marker='o', color='black', label='Best',s=100)
plt.text(0, .25, 'AUC: {}'.format(auc_), style='italic',
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})
# axis labels
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend()
# show the plot
plt.show()
```
